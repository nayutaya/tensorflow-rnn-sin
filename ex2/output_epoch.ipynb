```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
initial = np.load("lstm_basic/initial.npy")
initial_df = pd.DataFrame(initial, columns=["initial"])
```

```python
output1 = np.load("lstm_epoch1000/output.npy")
output1_df = pd.DataFrame(output1, columns=["output (epoch: 1000)"], index=range(len(initial), len(initial) + len(output1)))
losses1_df = pd.DataFrame(np.load("lstm_epoch1000/losses.npy"), columns=["epoch", "loss (epoch: 1000)"])
```

```python
output2 = np.load("lstm_basic/output.npy")
output2_df = pd.DataFrame(output2, columns=["output (epoch: 2000)"], index=range(len(initial), len(initial) + len(output2)))
losses2_df = pd.DataFrame(np.load("lstm_basic/losses.npy"), columns=["epoch", "loss (epoch: 2000)"])
```

```python
output3 = np.load("lstm_epoch3000/output.npy")
output3_df = pd.DataFrame(output3, columns=["output (epoch: 3000)"], index=range(len(initial), len(initial) + len(output3)))
losses3_df = pd.DataFrame(np.load("lstm_epoch3000/losses.npy"), columns=["epoch", "loss (epoch: 3000)"])
```

```python
train = np.load("train_data/normal.npy")
train_df = pd.DataFrame(train[:, 0], columns=["train"])
```

```python
merged = pd.concat([train_df, initial_df, output1_df, output2_df, output3_df])
merged.plot(figsize=(15, 5), grid=True, style=["-", "-", "-", "-", "k--"])
```

```python
merged = pd.merge(losses1_df, losses2_df, how="right")
merged = pd.merge(merged, losses3_df, how="right")
merged.plot(figsize=(15, 5), grid=True, logy=True, x="epoch")
```
