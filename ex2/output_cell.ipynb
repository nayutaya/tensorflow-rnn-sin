```python
import pandas as pd
import numpy as np
import yaml
%matplotlib inline
```

```python
initial = np.load("lstm_basic/initial.npy")
initial_df = pd.DataFrame(initial, columns=["initial"])
```

```python
output1 = np.load("lstm_basic/output.npy")
output1_df = pd.DataFrame(output1, columns=["output (cell: LSTM)"], index=range(len(initial), len(initial) + len(output1)))
losses1_df = pd.DataFrame(np.load("lstm_basic/losses.npy"), columns=["epoch", "loss (cell: LSTM)"])
```

```python
output2 = np.load("gru_basic/output.npy")
output2_df = pd.DataFrame(output2, columns=["output (cell: GRU)"], index=range(len(initial), len(initial) + len(output2)))
losses2_df = pd.DataFrame(np.load("gru_basic/losses.npy"), columns=["epoch", "loss (cell: GRU)"])
```

```python
train = np.load("train_data/normal.npy")
train_df = pd.DataFrame(train[:, 0], columns=["train"])
```

```python
merged = pd.concat([train_df, initial_df, output1_df, output2_df])
merged.plot(figsize=(15, 5), grid=True, style=["-", "-", "-", "k--"])
```

```python
merged = pd.merge(losses1_df, losses2_df)
merged.plot(figsize=(15, 5), grid=True, logy=True, x="epoch")
```
